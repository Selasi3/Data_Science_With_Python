```python
tips = pd.read_csv("../dataset/tips_data.csv")
```

```python
tips.info()
```

### Scatter Plot

```python
# creating a scatterplot with a low level function - sactterplot() 

sns.scatterplot(x='total_bill',y='tip',data = tips)
plt.show()
```

### Relational  Plot

```python
sns.relplot(x='total_bill',y = 'tip',data = tips,kind = 'scatter')
plt.show()
```

```python
# Specifying rows for subplots

sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           row ='sex' )
plt.show()
```

```python
# Specifying columns for subplots

sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
            col ='day'
           )
plt.show()
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           size = 'size')
plt.show()
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           size = 'size',
           hue = 'size')
plt.show()
```

```python
sns.relplot(x='total_bill',y = 'tip',data = tips,
            kind = 'scatter',
           hue = 'smoker',
           style= 'smoker')
plt.show()
```

```python
tips.head()
```

```python
sns.boxenplot(x="sex",
               y="size", 
               data=tips,
              palette='coolwarm')
```

```python
sns.catplot(x="sex",data=tips,kind='count')
plt.xticks(rotation =45)


# Show plot
plt.show()
```
