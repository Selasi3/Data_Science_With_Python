### csv


CSV  means Comma Separated Value.

It is a file format used for storing and exchaning tablular data, such as spreadsheets or databases, in a plain-text format.

Each line in a csv filerepresents a row of data, and each value within a row is separated by a comma (or other specified delimiter character). 

The first row of the file usually contains the column headers, while subsequent rows contain the actual data.
